```python
import ast
def date_to_age(bdate: str) -> int:
    date = bdate.split(".")
    if len(date) < 3:
        return None
    else:
        return 2020 - int(date[2])

def count_info(row: pd.Series) -> int:
    columns = ["activities", "interests", "music", "movies", "tv", "books", "about", "inspired_by", "religion"]
    return sum([ 1 if pd.isna(row[column]) else 0 for column in columns])

df.posts = df.posts.apply(lambda ps: np.array(ast.literal_eval(ps)))

df = df.assign( age = df.bdate.apply(date_to_age),
                posts_count = df.posts.apply(len),
                posts_len = df.posts.apply(lambda ps: int(np.median([len(p) for p in ps]) if len(ps) > 0 else 0)),
                info_count = df.apply(count_info, axis=1))
df = df[["age", "relation", "posts_count", "posts_len", "info_count", "sex"]]
df = df[df.age.notnull()]
df['sex'] = (df['sex'] == 'male').astype(int)
df.head()
```

```python
# –í–Ω–µ—à–Ω—è—è –º–µ—Ä–∞ ü§î
def create_adjacent(X, y, y_pred):
    assert len(y) == len(X) and len(X) == len(y_pred)

    n = np.zeros((np.max(y) + 1, np.max(y_pred) + 1))
    a = np.zeros(np.max(y) + 1)
    b = np.zeros(np.max(y_pred) + 1)
    for (yi, yi_pred) in zip(y, y_pred):
        n[yi][yi_pred] += 1
        a[yi] += 1
        b[yi_pred] += 1

    p = n / len(X)
    pi = a / len(X)
    pj = b / len(X)
    
    return (p, pi, pj)
def F(X, y, y_pred):
    p, pi, pj = create_adjacent(X, y, y_pred)
    s = 0
    for j in range(len(pj)):
        m = 0
        for i in range(len(pi)):
            if p[i][j] == 0 or pi[i] == 0 or pj[j] == 0:
                continue
            v = 2 * (p[i][j] / pi[i]) * (p[i][j] / pj[j]) / (p[i][j] / pi[i] + p[i][j] / pj[j])
            if v > m:
                m = v
        s += pj[j] * m
    return s
```

```python
print('F:', F(X, y, y_pred))
```

```python
def split_clasters(X, y_pred):
    res = [[] for _ in range(np.max(y_pred) + 1)]
    for (x, y) in zip(X.values, y_pred):
        res[y].append(x)
    return list(map(np.array, res))

def dist(a, b):
    return np.sqrt((a - b) @ (a - b))
```

```python
# –°—Ä–µ–¥–Ω–µ–µ –≤–Ω—É—Ç—Ä–∏–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ (–ö–æ–º–ø–∞–∫—Ç–Ω–æ—Å—Ç—å –∫–ª–∞—Å—Ç–µ—Ä–æ–≤)
def Cohension(X, y_pred):
    cl = split_clasters(X, y_pred)
    res = 0
    for c in cl:
        cm = c.mean(axis=0)
        for x in c:
            res += dist(cm, x)**2
    return res
```

```python
# –°—Ä–µ–¥–Ω–µ–µ –º–µ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ (–û—Ç–¥–µ–ª–∏–º–æ—Å—Ç—å –∫–ª–∞—Å—Ç–µ—Ä–æ–≤)
def Separation(X, y_pred):
    cl = split_clasters(X, y_pred)
    m = X.values.mean(axis=0)
    res = 0
    for c in cl:
        cm = c.mean(axis=0)
        res += dist(cm, m)**2
    return len(X) * res
```

```python
print("Cluster Cohension:", Cohension(X, y_pred))
print("Cluster Separation:", Separation(X, y_pred))
```

```python
from sklearn.cluster import KMeans, AffinityPropagation, MeanShift, SpectralClustering, AgglomerativeClustering, OPTICS, Birch
import time

clizers = [
    (KMeans(n_clusters=8, random_state=42), "KMeans"),
#     (AffinityPropagation(random_state=42), "Affinity Propagation"),
    (MeanShift(), "Mean shift"),
    (AgglomerativeClustering(n_clusters=8), "Agglomerative Clustering"),
    (OPTICS(), "OPTICS"),
    (Birch(n_clusters=8), "BIRCH"),
    (SpectralClustering(n_clusters=8, random_state=42), "Spectral Clustering"),
]

for (clizer, name) in clizers:
    start = time.time()
    y_pred = clizer.fit_predict(X_reduced)
    end = time.time()
    print(f'== {name} ==')
    print(f'  F-Metric:   {F(X, y, y_pred)}')
    print(f'  Cohension:  {Cohension(X, y_pred)}')
    print(f'  Separation: {Separation(X, y_pred)}')
    print(f'  Time:       {end - start} sec')
```
