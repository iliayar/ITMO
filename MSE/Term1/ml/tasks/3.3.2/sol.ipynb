```python
import numpy as np
import pandas as pd

# ----- TRAIN DATA -----

# число писем
n_spam = 20
n_ham = 26

# общее число слов
words_spam = 157
words_ham = 116

# таблица слов
data = {
    "word": ["Investment","Membership","Access","Bill","Online","Bonus",
             "Prize","Free","Remove","Refund"],
    "spam":  [1,2,0,5,8,0,19,10,44,68],
    "ham":   [3,3,0,2,0,10,13,23,4,58]
}

df = pd.DataFrame(data)

# ----- PRIORS -----
P_spam = n_spam / (n_spam + n_ham)
P_ham  = n_ham  / (n_spam + n_ham)

print("P(spam) =", P_spam)
print("P(ham)  =", P_ham)

# ----- LAPLACE SMOOTHING -----
V = len(df) # количество уникальных слов

def log_likelihood(message_words, cls):
    """cls = 'spam' or 'ham'"""
    if cls == "spam":
        total = words_spam
        counts = df.set_index("word")["spam"]
    else:
        total = words_ham
        counts = df.set_index("word")["ham"]

    ll = np.log((n_spam if cls == "spam" else n_ham) / (n_spam + n_ham))
    for w in message_words:
        c = counts[w] if w in counts.index else 0
        ll += np.log((c + 1) / (total + V + 2))
    return ll

# ----- MESSAGE -----

message = "Remove Million Bonus Access Online Money Bill".split()

# ----- COMPUTE F(spam) and F(ham) -----

F_spam = log_likelihood(message, "spam")
F_ham  = log_likelihood(message, "ham")

print("F(spam) =", F_spam)
print("F(ham)  =", F_ham)

# ----- POSTERIOR -----

log_posterior_spam = F_spam
log_posterior_ham  = F_ham

# нормировка
posterior_spam = np.exp(log_posterior_spam) / (np.exp(log_posterior_spam) + np.exp(log_posterior_ham))

print("P(Class=spam | message) =", posterior_spam)
```
