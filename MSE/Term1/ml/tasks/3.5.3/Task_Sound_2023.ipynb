–î–Ψ–≤–Ψ–Μ―¨–Ϋ–Ψ –Ω―Ä–Ψ―¹―²–Ψ–Ι –Κ–Ψ–¥ –¥–Μ―è ―Ä–Α―¹–Ω–Ψ–Ζ–Ϋ–Α–≤–Α–Ϋ–Η―è –Ζ–≤―É–Κ–Ψ–≤―΄―Ö ―³–Α–Ι–Μ–Ψ–≤ –Η –Ψ–±―Ä–Α–±–Ψ―²–Κ–Η –≤–Ψ–Ζ–Φ–Ψ–Ε–Ϋ―΄―Ö –Ψ―à–Η–±–Ψ–Κ:

```python
# –‰–Φ―è ―³–Α–Ι–Μ–Α, –Κ–Ψ―²–Ψ―Ä―΄–Ι ―²―Ä–Β–±―É–Β―²―¹―è ―Ä–Α―¹–Ω–Ψ–Ζ–Ϋ–Α―²―¨
AUDIO_FILE = '12.wav'

# –ß―²–Β–Ϋ–Η–Β ―³–Α–Ι–Μ–Α
r = sr.Recognizer()
with sr.AudioFile(AUDIO_FILE) as source:
    audio = r.record(source)

# –†–Α―¹–Ω–Ψ–Ζ–Ϋ–Α–≤–Α–Ϋ–Η–Β ―¹ –Ω–Ψ–Φ–Ψ―â―¨―é Sphinx
try:
    print("–†–Β–Ζ―É–Μ―¨―²–Α―² Sphinx: " + r.recognize_sphinx(audio))
except sr.UnknownValueError:
    print("–û―à–Η–±–Κ–Α: Sphinx could not understand audio")
except sr.RequestError as e:
    print("–û―à–Η–±–Κ–Α: Sphinx error; {0}".format(e))
```
