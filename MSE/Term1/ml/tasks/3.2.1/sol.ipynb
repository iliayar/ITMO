```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("94_16.csv", header=None)
```

```python
pca = PCA(svd_solver='full')
Z = pca.fit_transform(data)
```

```python
pc1_obj1 = Z[0, 0]   # относительно первой ГК
pc2_obj1 = Z[0, 1]   # относительно второй ГК
print("PC1 первого объекта:", pc1_obj1)
print("PC2 первого объекта:", pc2_obj1)
```

```python
expl2 = pca.explained_variance_ratio_[:2].sum()
print("Доля объяснённой дисперсии 2 компонентами:", expl2)
```

```python
cum = np.cumsum(pca.explained_variance_ratio_)
k85 = np.argmax(cum > 0.85) + 1
print("Минимальное число компонент для >0.85:", k85)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Z[:, 0], Z[:, 1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Объекты в пространстве первых двух главных компонент")
plt.grid(True)
plt.show()
```
