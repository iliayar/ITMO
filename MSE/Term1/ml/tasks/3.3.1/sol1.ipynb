````python
import numpy as np

# Новый объект
new_point = np.array([92, 85])

# Вычисляем евклидовы расстояния до всех точек
distances = np.sqrt((df['X'] - new_point[0])**2 + (df['Y'] - new_point[1])**2)

# Находим минимальное расстояние и соответствующую точку
min_distance = distances.min()
nearest_idx = distances.idxmin()

print(f"Ближайший сосед: точка {df.loc[nearest_idx, 'id']} с координатами ({df.loc[nearest_idx, 'X']}, {df.loc[nearest_idx, 'Y']})")
print(f"Расстояние: {min_distance}")
print(f"\nВсе расстояния:")
print(distances.sort_values())
````

````python
# Находим 3 ближайших соседа
k_nearest = distances.nsmallest(3)

print("3 ближайших соседа:")
for idx in k_nearest.index:
    print(f"ID: {df.loc[idx, 'id']}, координаты: ({df.loc[idx, 'X']}, {df.loc[idx, 'Y']}), расстояние: {distances[idx]:.2f}")

print(f"\nИдентификаторы трех ближайших точек: {list(df.loc[k_nearest.index, 'id'].values)}")
````

````python
# Получаем классы 3 ближайших соседей
k = 3
k_nearest_indices = distances.nsmallest(k).index
nearest_classes = df.loc[k_nearest_indices, 'Class'].values

print("Классы 3 ближайших соседей:")
for idx in k_nearest_indices:
    print(f"ID: {df.loc[idx, 'id']}, класс: {df.loc[idx, 'Class']}")

# Определяем класс методом голосования (большинство голосов)
from collections import Counter
class_counts = Counter(nearest_classes)
predicted_class = class_counts.most_common(1)[0][0]

print(f"\nКоличество голосов по классам: {dict(class_counts)}")
print(f"Предсказанный класс для точки (92, 85): {predicted_class}")
````
