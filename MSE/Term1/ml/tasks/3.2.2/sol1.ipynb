```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data.csv")
```

```python
X = df[["X"]]
y = df["Y"]
```

```python
mean_X = df["X"].mean()
mean_Y = df["Y"].mean()
```

```python
model = LinearRegression().fit(X, y)
```

```python
theta1 = model.coef_[0]
theta0 = model.intercept_
```

```python
r2 = model.score(X, y)
```

```python
print("mean X:", mean_X)
print("mean Y:", mean_Y)
print("theta1:", theta1)
print("theta0:", theta0)
print("R2:", r2)
```
