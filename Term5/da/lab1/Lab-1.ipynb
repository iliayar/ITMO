![relation](relation.png)

```python
def date_to_age(bdate: str) -> int:
    from datetime import datetime
    
    date = bdate.split(".")
    if len(date) < 3:
        return None
    else:
        return datetime.now().year - int(date[2])

def count_info(row: pd.Series) -> int:
    columns = ["activities", "interests", "music", "movies", "tv", "books", "about", "inspired_by", "religion"]
    return sum([ 1 if pd.isna(row[column]) else 0 for column in columns])
    
df = df1.assign( age = df1.bdate.apply(date_to_age),
                posts_count = df1.posts.apply(len),
                info_count = df1.apply(count_info, axis=1))
df = df[["age", "relation", "posts_count", "info_count"]]
df = df[df.age.notnull()]
df.head()
```

```python
from scipy import interpolate
import numpy as np

N = 10
XMAX = 10
XMIN = -10
YMAX = 10
YMIN = -10
```

```python
from typing import List, Tuple, Union

def my_interpolate(xs: List[float], ys: List[float]):
    xs = np.array(xs)
    ys = np.array(ys)
    
    hs = np.diff(xs)
    diffs = np.diff(ys) / hs
    
    d_signs = np.sign(diffs)
    zeros = (d_signs[1:] != d_signs[:-1]) | (diffs[1:] == 0) | (diffs[:-1] == 0)
    
    w1s = 2 * hs[1:] + hs[:-1]
    w2s = hs[1:] + 2 * hs[:-1]
    
    wmeans = (w1s + w2s) / (w1s / diffs[:-1] + w2s / diffs[1:])
    
    derivs = np.empty(xs.shape, dtype=xs.dtype)
    derivs[1:-1][zeros]  = 0.0
    derivs[1:-1][~zeros] = wmeans[~zeros]
    
    t = (derivs[1:] + derivs[:-1] - 2 * diffs) / hs
    
    koefs = np.empty((4,) + hs.shape, dtype=t.dtype)
    koefs[0] = t / hs
    koefs[1] = (diffs - derivs[:-1]) / hs - t
    koefs[2] = derivs[:-1]
    koefs[3] = ys[:-1]
        
    def evaluate(x: float):
        point = xs[xs < x].shape[0]
        rank = koefs.shape[0] - 1
        
        if point == xs.shape[0]:
            interval = koefs.shape[1] - 1
            point = xs.shape[0] - 1
        elif point == 0:
            point = 0
            interval = 0
        else:
            point -= 1
            interval = point
        
        return koefs[:, interval] @ (x - xs[point])**(rank - np.arange(rank + 1))
    
    return evaluate
```

```python
# –ü—Ä–∏–º–µ—Ä –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã —Å –ø–æ–º–æ—â—å—é plt.hist()
import matplotlib.pyplot as plt

# array = [1,1,2,3,4,1,2,3,4]
array = np.random.randint(0, 60, (50,))

value_counts, bins_names, visual_figure = plt.hist(array, 
                                                   bins=len(set(array))
                                                  ,rwidth = 0.6)
# –ü–∞—Ä–∞–º–µ—Ç—Ä rwidth –∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é –≤ —Å–≤–æ–µ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–µ–∞–ª–∏–∑–æ–≤—ã–≤–∞—Ç—å –Ω–µ —Ç—Ä–µ–±—É–µ—Ç—Å—è
print('–ó–Ω–∞—á–µ–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫:', value_counts)
print('–ù–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫:', bins_names)
```

```python
from typing import List, Tuple, Union

def fast_hist(array: List[Union[int, float]], 
              bins: int) -> Tuple[List[int], List[float]]:
    """
    Builds bins' labels and bins' value counts for given array
    :param array: array with numeric values
    :param bins:  number of bins in result distribution
    :return: Two lists: 
             first contains value counts of each bin,
             second contains list of bins' labels
    """
    
    bins_cnt = bins
    array.sort()
    a_min, a_max = array[0], array[-1]
    step = (a_max - a_min) / bins_cnt
    
    cur = a_min + step
    
    values = [0]
    bins = [a_min, cur]
    
    i = 0
    for e in array:
        while e > cur:
            i += 1
            if i >= bins_cnt:
                cur = a_max
            else:
                cur += step
            bins.append(cur)
            values.append(0)
        values[-1] += 1
    
    return (values, bins)
def bins_to_names(bins: List[float]):
    prev = bins[0]
    nbins = []
    for e in bins[1:]:
        nbins.append((prev + e)/2)
        prev = e
    return nbins
```

```python
value_counts, bins = fast_hist(array, bins=len(set(array)))
plt.bar(bins_to_names(bins), value_counts)

print('–ó–Ω–∞—á–µ–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫:', value_counts)
print('–ù–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫:', bins)
```

```python
def plot_lib(array, bins):
    value_counts, bins_names, visual_figure = plt.hist(array, 
                                                   bins=len(set(array)), 
                                                   rwidth=1)
def plot_my(array, bins):
    value_counts, bins = fast_hist(array, bins=len(set(array)))
    plt.bar(bins_to_names(bins), value_counts)
```
